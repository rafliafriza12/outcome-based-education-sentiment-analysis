```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data performa model
data = {
    "Model": ["Fine-tuned IndoBERT", "Naive Bayes", "SVM"],
    "Accuracy": [0.74, 0.47, 0.55],
    "ROC AUC": [0.859, 0.627, 0.754],
    "F1-Score": [0.73, 0.47, 0.55],
    "precission": [0.74, 0.49, 0.58],
    "recall": [0.75, 0.47, 0.55]
}
```

```python
# Konversi ke DataFrame
df = pd.DataFrame(data)
```

```python
# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x="Model", y="Accuracy", data=df, label="Accuracy", alpha=0.7)
sns.barplot(x="Model", y="ROC AUC", data=df, label="ROC AUC", alpha=0.7)
sns.barplot(x="Model", y="F1-Score", data=df, label="Macro F1-Score", alpha=0.7)
sns.barplot(x="Model", y="precission", data=df, label="precission", alpha=0.7)
sns.barplot(x="Model", y="recall", data=df, label="recall", alpha=0.7)

plt.xlabel("Model")
plt.ylabel("Score")
plt.title("Perbandingan Performa Model")
plt.legend()
plt.ylim(0, 1)
plt.show()
```

```python
# Plot
plt.figure(figsize=(10, 6))
df.set_index("Model").plot(kind="bar", figsize=(10, 6), colormap="viridis")

plt.xlabel("Model")
plt.ylabel("Score")
plt.title("Perbandingan Performa Model")
plt.ylim(0, 1)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
